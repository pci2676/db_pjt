여기서 필요한 정보를 제외하고는 삭제를 진행

```python
#delete the non_valueable
del total_data['age']
del total_data['total_m']
del total_data['total_w']
del total_data['in_m']
del total_data['in_w']
del total_data['out_m']
del total_data['out_w']
total_data.head()
```

### Correlation Analysis

상관분석은 numpy 모듈의 corrcoef를 사용함

상관분석에서

절댓값 0.1 이하는 무시하고

절댓값 0.3 이하는 약한 상관관계를

절댓값 0.7 이하는 강한 상관관계를 나타낸다

우선은 cctv와 교통사고사망자, 소득, 범죄율, 범죄수, 자살율, 외국인 비율과의 상관관계를 분석

```python
print('cctv / 교통사고사망자: ')
print(np.corrcoef(total_data['CCTV'], total_data['교통사망자']))
print('cctv / 소득: ')
print(np.corrcoef(total_data['CCTV'], total_data['tax']))
print('cctv / 범죄율: ')
print(np.corrcoef(total_data['CCTV'], total_data['crimeRate']))
print('cctv / 범죄수: ')
print(np.corrcoef(total_data['CCTV'], total_data['범죄수']))
print('cctv / 자살율: ')
print(np.corrcoef(total_data['CCTV'], total_data['suicide_rate']))
print('cctv / 외국인비율: ')
print(np.corrcoef(total_data['CCTV'], total_data['out_per']))
```

그 다음은 인구수와 상관관계를 보려고 한다

```python
print('인구수 / 교통사고사망자: ')
print(np.corrcoef(total_data['total'], total_data['교통사망자']))
print('인구수 / 소득: ')
print(np.corrcoef(total_data['total'], total_data['tax']))
print('인구수 / 범죄율: ')
print(np.corrcoef(total_data['total'], total_data['crimeRate']))
print('인구수 / 범죄수: ')
print(np.corrcoef(total_data['total'], total_data['범죄수']))
print('인구수 / 자살율: ')
print(np.corrcoef(total_data['total'], total_data['suicide_rate']))
print('인구수 / 외국인비율: ')
print(np.corrcoef(total_data['total'], total_data['out_per']))
```
